```python
def read_csfastq(ruta):
    
    with open(ruta, "r") as file:
        
        while True:
            
            id_line = file.readline().strip()
            
            if not id_line:
                break
            
            seq_line = file.readline().strip()
            plus_line = file.readline().strip()
            qual_line = file.readline().strip()

            yield id_line, seq_line, plus_line, qual_line
```

```python
def convierte_secs(ruta):
    
    dicc_convers = {"T0": "T", "T1": "G", "T2": "C", "T3": "A", "T.": "N",
        "C0": "C", "C1": "A", "C2": "T", "C3": "G", "C.": "N",
        "G0": "G", "G1": "T", "G2": "A", "G3": "C", "G.": "N",
        "A0": "A", "A1": "C", "A2": "G", "A3": "T", "A.": "N",
        "N0": "N", "N1": "N", "N2": "N", "N3": "N", "N.": "N"}

    secuencias = []
    for salida in read_csfastq(ruta):

        seq_csfastq = salida[1]

        base_actual = seq_csfastq[0]

        seq_fastq = [base_actual]

        for pos in seq_csfastq[1:]:
            base = dicc_convers[str(base_actual)+str(pos)]
            seq_fastq.append(base)
            base_actual = base

        secuencia_fastq = ''.join(seq_fastq)
        secuencias.append(secuencia_fastq)
        
    return secuencias
```

```python
secs_fastq = convierte_secs(r"C:\Users\34601\Downloads\SRR922052.fastq")
```
